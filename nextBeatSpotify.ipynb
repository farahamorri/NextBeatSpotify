### 1. Configuration et chargement

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import kagglehub
import os

# Configuration de l'affichage pour voir toutes les colonnes
pd.set_option('display.max_columns', None)

# 1. Chargement du dataset
# 1. Téléchargement du dataset (dernière version)
path = kagglehub.dataset_download("joebeachcapital/30000-spotify-songs")
print("✅ Dataset téléchargé à l'adresse :", path)

# 2. Lister les fichiers pour trouver le CSV
files = os.listdir(path)
csv_file = [f for f in files if f.endswith('.csv')][0]
full_path = os.path.join(path, csv_file)

# 3. Chargement dans un DataFrame Pandas
df = pd.read_csv(full_path)
# 2. Premier aperçu des données
print("\n--- Aperçu des 5 premières lignes ---")
display(df.head())

# 3. Structure du dataset (Nombre de lignes/colonnes et types)
print("\n--- Informations sur le dataset ---")
df.info()

# 4. Vérification des statistiques de base
print("\n--- Statistiques descriptives ---")
display(df.describe())
```

### 2. Nettoyage et exploration

```python
# 1. Suppression des doublons
# Dans les datasets Spotify, une chanson peut apparaître plusieurs fois si elle est dans plusieurs albums.
print(f"Nombre de lignes avant : {len(df)}")
df = df.drop_duplicates(subset=['track_name', 'track_artist'])
print(f"Nombre de lignes après suppression des doublons : {len(df)}")

# 2. Nettoyage de la colonne 'track_artist' (pour enlever les crochets et guillemets)
# La colonne est souvent sous forme de chaîne "[ 'Artiste' ]"
df['track_artist'] = df['track_artist'].str.replace("[", "").str.replace("]", "").str.replace("'", "")

# 3. Sélection des caractéristiques audio (Features)
# On ne garde que les colonnes numériques qui décrivent le "son"
audio_features = [
    'danceability', 'energy', 'key', 'loudness', 'mode', 
    'speechiness', 'acousticness', 'instrumentalness', 
    'liveness', 'valence', 'tempo'
]

# On crée un nouveau DataFrame avec uniquement ces colonnes pour l'analyse
df_features = df[audio_features]

# 4. Vérification des valeurs manquantes
print("\nValeurs manquantes par colonne :")
print(df_features.isnull().sum())

# 5. Visualisation : Matrice de Corrélation
# Cela permet de voir si certaines caractéristiques sont liées (ex: Energy et Loudness)
plt.figure(figsize=(12, 8))
sns.heatmap(df_features.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Corrélation entre les caractéristiques audio")
plt.show()

# 6. Distribution d'une feature (Exemple : Danceability)
plt.figure(figsize=(10, 4))
sns.histplot(df['danceability'], kde=True, color='green')
plt.title("Distribution de la Danceability dans le dataset")
plt.show()
```

### 3. Prétraitement: la Normalisation

```python
from sklearn.preprocessing import MinMaxScaler

# 1. Initialisation du Scaler
# On utilise MinMaxScaler pour ramener toutes les valeurs entre 0 et 1.
# Cela préserve la distribution originale tout en égalisant l'importance de chaque feature.
scaler = MinMaxScaler()

# 2. Application de la normalisation
# Nous appliquons le scaler sur notre DataFrame 'df_features' créé à l'étape précédente.
scaled_features = scaler.fit_transform(df_features)

# 3. Création d'un DataFrame avec les données normalisées
# On conserve les noms de colonnes et l'index original pour pouvoir retrouver les chansons plus tard.
df_scaled = pd.DataFrame(scaled_features, columns=audio_features, index=df.index)

# 4. Vérification du résultat
print("--- Aperçu des données normalisées ---")
display(df_scaled.head())

# Vérification que toutes les valeurs sont bien entre 0 et 1
print("\nStatistiques après normalisation (Min/Max) :")
print(f"Minimum global : {df_scaled.min().min()}")
print(f"Maximum global : {df_scaled.max().max()}")
```

### 4. Construction de l'Autoencoder 

```python
import tensorflow as tf
from tensorflow.keras import layers, models

# 1. Définition de la dimension d'entrée (le nombre de features audio)
input_dim = df_scaled.shape[1] 
latent_dim = 4  # On compresse les 11 features en 4 "super-features"

# 2. L'ENCODEUR
input_layer = layers.Input(shape=(input_dim,))
encoded = layers.Dense(8, activation='relu')(input_layer)
bottleneck = layers.Dense(latent_dim, activation='relu')(encoded)

# 3. LE DÉCODEUR
decoded = layers.Dense(8, activation='relu')(bottleneck)
output_layer = layers.Dense(input_dim, activation='sigmoid')(decoded)

# 4. LE MODÈLE COMPLET (Autoencoder)
autoencoder = models.Model(inputs=input_layer, outputs=output_layer)

# 5. L'ENCODEUR SEUL (Pour la recommandation)
encoder = models.Model(inputs=input_layer, outputs=bottleneck)

autoencoder.compile(optimizer='adam', loss='mse')
autoencoder.summary()
```

### 5. Entrainement du modèle 

```python
# Entraînement
history = autoencoder.fit(
    df_scaled, df_scaled,
    epochs=50,
    batch_size=32,
    shuffle=True,
    validation_split=0.1,
    verbose=1
)

# Visualisation de la perte
plt.plot(history.history['loss'], label='Perte Entraînement')
plt.plot(history.history['val_loss'], label='Perte Validation')
plt.legend()
plt.title("Évolution de l'apprentissage")
plt.show()
```

### 6. Génération des recommandations

```python
from sklearn.metrics.pairwise import cosine_similarity

# 1. On compresse toutes les chansons dans l'espace latent
latent_vectors = encoder.predict(df_scaled)

def recommend_with_autoencoder(song_name, df, latent_vectors, n_recs=5):
    if song_name not in df['track_name'].values:
        return "Chanson non trouvée."

    # Index de la chanson
    idx = df[df['track_name'] == song_name].index[0]
    
    # Vecteur latent de la chanson cible
    target_vector = latent_vectors[idx].reshape(1, -1)
    
    # Similarité dans l'espace latent
    similarity = cosine_similarity(target_vector, latent_vectors)
    
    # Top indices
    similar_indices = similarity[0].argsort()[::-1][1:n_recs+1]
    
    return df.iloc[similar_indices][['track_name', 'track_artist', 'track_album_release_date']]

# TEST
print("Recommandations via Autoencoder :")
display(recommend_with_autoencoder(df['track_name'].iloc[0], df, latent_vectors))
```
